```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df['Date_of_Journey'].str.split('/')
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Dep_Time']
```

```python
df['Dep_Time'].str.split(":")
```

```python
df['Hour']=df['Dep_Time'].str.split(":").str[0]
df['Minutes']=df['Dep_Time'].str.split(":").str[1]
```

```python
df.head()
```

```python
df['Dep_Hour']=df['Hour'].astype(int)
df['Dep_Minutes']=df['Minutes'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Arrival_Time']
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0].astype(int)
df['Arrival_Min']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Duration']
```

```python
df['Duration'].str.split(' ')
```

```python
df['Duration_Hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0].str.replace('m','').str[0].astype(int)
df['Duration_Min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0].astype(float)
```

```python
df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.drop('Duration',axis=1,inplace=True)
```
